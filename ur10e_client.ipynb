```python
import numpy as np
import json

import requests
from PIL import Image
import io
```

```python
url = "http://127.0.0.1:5555"
```

# Stop Simulation

```python
response = requests.get(f"{url}/stop")
response.status_code
```

# FK & IK

```python
# ee_position 위치 조회
response = requests.get(f"{url}/get_ee")
response.status_code
data = response.json()
data
```

```python
# obj_position 위치 조회
headers = {'Content-type': 'application/json'}
cmd = {"site_name": "table_x_1_y_1"}
response = requests.post(f"{url}/get_site", headers=headers, data=json.dumps(cmd))
response.status_code
data = response.json()
data
```
